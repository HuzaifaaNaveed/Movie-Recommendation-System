```python
import pandas as pd
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
# @title Movies Histogram
movies.hist(figsize=(10, 10),color='salmon')
```

```python
# @title Budget vs Revenue Scatterplot of Movies
plt.figure(figsize=(8, 6))
sns.scatterplot(x='budget', y='revenue', data=movies)
plt.title('Budget vs. Revenue')
plt.show()
```

```python
# @title Correlation Matrix of Numeric Movie Features
numeric=movies.select_dtypes(include=['number'])
corr=numeric.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='viridis')
plt.title('Correlation Matrix of Numeric Features of Movies')
plt.show()
```

```python
# @title Movie Count based on Original Language
cc=movies['original_language'].value_counts()
plt.figure(figsize=(12, 6))
ax=sns.barplot(x=cc.index,y=cc.values,color='orange')
plt.title('Movie Count by Original Language')
plt.xlabel('Original Language')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45,ha='right')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.show()
```

```python
# @title Cast Size Frequency Histplot
import ast
def process_cast(cast_str):
    cast_list = ast.literal_eval(cast_str)
    return len(cast_list)
credits['cast_size']=credits['cast'].apply(process_cast)
plt.figure(figsize=(8, 6))
sns.histplot(credits['cast_size'], kde=True, color='green')
plt.title('Distribution of Cast Size')
plt.xlabel('Number of Cast Members')
plt.ylabel('Frequency')
plt.show()
```

```python
# @title Crew Size Frequency Histplot
def process_crew(crew_str):
  crew_list = ast.literal_eval(crew_str)
  return len(crew_list)
credits['crew_size'] = credits['crew'].apply(process_crew)
plt.figure(figsize=(8, 6))
sns.histplot(credits['crew_size'], kde=True, color='purple')
plt.title('Distribution of Crew Size')
plt.xlabel('Number of Crew Members')
plt.ylabel('Frequency')
plt.show()
```

```python
# @title Movie ID vs Cast Size Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='movie_id', y='cast_size', data=credits, color='lightgreen')
plt.title('Movie ID vs. Cast Size')
plt.show()
```

```python
# @title Movie ID vs Crew Size Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='movie_id', y='crew_size', data=credits, color='grey')
plt.title('Movie ID vs. Crew Size')
plt.show()
```

```python
# @title Movie Genre Counts
g=[]
for gl in movies['genres']:
    for genre in gl:
        g.append(genre['name'])
df=pd.DataFrame(g,columns=['genre'])
plt.figure(figsize=(12, 6))
sns.countplot(y='genre', data=df, order=df['genre'].value_counts().index, color='purple')
plt.title('Distribution of Movie Genres')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

```python
# @title Top 10 Production Companies
plt.figure(figsize=(12, 6))
p=[]
for index in movies.index:
    for c in movies.loc[index,'production_countries']:
        p.append(c['name'])
pdf=pd.DataFrame(p, columns=['production_country'])
pdf=pdf[pdf['production_country'] != '']
sns.countplot(y='production_country',data=pdf,order=pdf['production_country'].value_counts().index[:10],color='orange')
plt.title('Top 10 Production Countries')
plt.xlabel('Number of Movies')
plt.ylabel('Production Country')
plt.show()
```

```python
# @title Popularity vs Vote Average Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='popularity', y='vote_average', data=movies,color='yellow')
plt.title('Popularity vs. Vote Average')
plt.show()
```

```python
# @title Runtime vs Revenue Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='runtime', y='revenue', data=movies,color='black')
plt.title('Runtime vs. Revenue')
plt.show()
```

```python
# @title Boxplot of Original Language w.r.t Popularity
plt.figure(figsize=(15,10))
sns.boxplot(x='original_language', y='popularity', data=movies)
plt.title('Revenue Distribution by Original Language')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# @title Popularity Based on Genre
movies=movies.merge(credits, on='title')
a=[]
for index in movies.index:
    for genre in movies.loc[index, 'genres']:
        a.append([genre['name'], movies.loc[index,'popularity']])
df=pd.DataFrame(a,columns=['genre', 'popularity'])
avg=df.groupby('genre')['popularity'].mean()
avg=avg.sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg.values, y=avg.index, color='skyblue')
plt
```

```python
# @title Relationship between Budget and Popularity
plt.figure(figsize=(8, 6))
sns.regplot(x='budget', y='popularity', data=movies,color='brown')
plt.title('Budget vs. Popularity')
plt.show()
```

```python
movies['release_date']=pd.to_datetime(movies['release_date'], errors='coerce')
movies['release_year']=movies['release_date'].dt.year
plt.figure(figsize=(12, 6))
sns.countplot(x='release_year', data=movies,color='green')
plt.title('Number of Movies Released Each Year')
plt.xticks(rotation=90)
plt.show()
```

```python
avg=movies.groupby('release_year')['revenue'].mean()
plt.figure(figsize=(12, 6))
sns.lineplot(x=avg.index, y=avg.values,color='magenta')
plt.title('Average Revenue per Year')
plt.xlabel('Year')
plt.ylabel('Average Revenue')
plt.show()
```

```python
# @title Relationship between Revenue and Popularity
plt.figure(figsize=(8, 6))
sns.regplot(x='revenue', y='popularity', data=movies)
plt.title('Revenue vs. Popularity')
plt.show()
```

```python
# @title Relationship between Budget and Popularity
plt.figure(figsize=(8, 6))
sns.regplot(x='budget', y='popularity', data=movies)
plt.title('Budget vs. Popularity')
plt.show()
```

```python
# @title Average Popularity per Year Lineplot
avg=movies.groupby('release_year')['popularity'].mean()
plt.figure(figsize=(12, 6))
sns.lineplot(x=avg.index, y=avg.values,color='orange')
plt.title('Average Popularity per Year')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.show()
```

```python
# @title Average Vote Average per Year Lineplot
avg_vote = movies.groupby('release_year')['vote_average'].mean()
plt.figure(figsize=(12, 6))
sns.lineplot(x=avg_vote.index, y=avg_vote.values, color='purple')
plt.title('Average Vote Average per Year')
plt.xlabel('Year')
plt.ylabel('Average Vote Average')
plt.show()
```

```python
# @title Average Vote Count per Year Lineplot
avg_vote_count = movies.groupby('release_year')['vote_count'].mean()
plt.figure(figsize=(12, 6))
sns.lineplot(x=avg_vote_count.index, y=avg_vote_count.values, color='red')
plt.title('Average Vote Count per Year')
plt.xlabel('Year')
plt.ylabel('Average Vote Count')
plt.show()
```

```python
# @title Movie Overview WordCloud
from wordcloud import WordCloud
text = " ".join(movies['overview'].astype(str))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Movie Overviews")
plt.show()
```

```python
# @title Movie Overview Sentiment Polarity
from textblob import TextBlob
movies['sentiment_polarity'] = movies['overview'].astype(str).apply(lambda x: TextBlob(x).sentiment.polarity)
plt.figure(figsize=(8, 6))
sns.histplot(movies['sentiment_polarity'], kde=True)
plt.title('Distribution of Sentiment Polarity in Movie Overviews')
plt.show()
```

```python
# @title Movie Overview Length
movies['overview_length']=movies['overview'].astype(str).apply(len)
plt.figure(figsize=(10, 6))
sns.histplot(movies['overview_length'], kde=True, color='skyblue')
plt.title('Distribution of Overview Length')
plt.xlabel('Overview Length (Characters)')
plt.ylabel('Frequency')
plt.show()
```
